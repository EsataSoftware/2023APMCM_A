```python
import os
import re

def rename_files(directory_path):
    # 获取目录中的文件列表
    file_list = os.listdir(directory_path)

    # 正则表达式模式，匹配包含下划线的文件名
    pattern = re.compile(r'(_)\((\d+)\)')

    for file_name in file_list:
        # 构建文件的完整路径
        file_path = os.path.join(directory_path, file_name)

        # 检查文件是否是一个普通文件而不是目录
        if os.path.isfile(file_path):
            # 进行正则表达式替换
            new_file_name = re.sub(pattern, r'(\2)', file_name)
            new_file_path = os.path.join(directory_path, new_file_name)

            # 重命名文件
            os.rename(file_path, new_file_path)
            print(f"重命名成功：{file_name} -> {new_file_name}")

# 用法示例，将路径替换为你的目录路径
rename_files('./DataCopyQuestion3/')
```

```python
import os

# 定义目录和txt文件路径
directory_path = "./DataCopyQuestion3/"
txt_file_path = './AppleClassifyNames.txt'

# 读取txt文件中的文件名
with open(txt_file_path, 'r') as file:
    valid_file_names = set(line.strip() for line in file)

# 遍历目录中的文件
for filename in os.listdir(directory_path):
    file_path = os.path.join(directory_path, filename)

    # 如果文件名不在txt文件中，删除文件
    if filename not in valid_file_names:
        os.remove(file_path)
        print(f"Deleted: {file_path}")
```
